## Day002 作業內容

我認為機器學習中的[增強式學習]是一個很值得發展的領域

原因如下：

１．他能夠在未知環境之下去執行，雖然第一次要去定義環境執行的步驟與時間需要花比較長的時間，但卻比較適合真實的世界。

２．透過Agent協助人類做出決策，並針對現在的狀態做出行動，以達到獲取較高的獎勵。而在行為過後，改變現在的環境狀態、獎勵變更後再回饋給代理人，已決定下一步的動作。這樣模式比較能應付問題中有比較多的不確定因素。

３．再透過Q-learning可以把問題延伸到全域解，而不是只有局部的解決問題。

根據以上3點原因，選擇增強式學習比較能應用在實體上面，如機器人動作、或是遊戲類等等的，比起監督式學習來得更能適合實際情況。
